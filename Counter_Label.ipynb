```python
import csv
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# Pfad zur CSV-Datei
csv_file_path = '/content/bereinigte_datei.csv'
```

```python
# Spaltenname, der gezählt werden soll
column_name = 'label'

# Zähler initialisieren
label_counter = Counter()

# CSV-Datei öffnen und verarbeiten
with open(csv_file_path, mode='r', newline='', encoding='utf-8') as file:
    reader = csv.DictReader(file)

    # Durch die Zeilen iterieren und die Werte der Spalte 'label' zählen
    for row in reader:
        label_value = row[column_name]
        label_counter[label_value] += 1
```

```python
# Anzahl der 0er, 1er und 2er ausgeben
count_0 = label_counter.get('0', 0)  # neutral
count_1 = label_counter.get('1', 0)  # negative
count_2 = label_counter.get('2', 0)  # positive

print(f"Anzahl der neutralen Posts: {count_0}")
print(f"Anzahl der negativen Posts: {count_1}")
print(f"Anzahl der positiben Posts: {count_2}")

# Gesamtsumme berechnen und ausgeben
total_sum = count_0 + count_1 + count_2
print(f"Gesamtsumme aller: {total_sum}")
```

```python
# Balkendiagramm erstellen
labels = ['Neutral', 'Negativ', 'Positiv']
counts = [count_0, count_1, count_2]

plt.figure(figsize=(8, 6))
bars = plt.bar(labels, counts, color=['#90EE90', '#FF7F7F', '#87CEEB'], edgecolor='black')

# Werte auf die Balken schreiben
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height / 2, f'{int(height)}', ha='center', va='center', color='black', fontsize=12)

# Diagramm beschriften
plt.xlabel('Sentiment', fontsize=14)
plt.ylabel('Anzahl der Posts', fontsize=14)
plt.title('Anzahl der Sentiments (Positiv, Negativ, Neutral)', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Diagramm anzeigen
plt.tight_layout()
plt.show()
```
